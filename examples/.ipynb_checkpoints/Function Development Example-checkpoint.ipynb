```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
```

```python
with rasterio.open('../sample_data/sample_dem.tif') as src:
    Z = src.read(1)
    Zt = src.transform
    cellsize = Zt[0]
    profile = src.profile

plt.imshow(Z,cmap='terrain',vmin=-500,vmax=2000)
plt.show()
```

```python
# Develop a function

def my_function(X,cellsize):
    X = np.gradient(X/cellsize)
    X = X[0]
    return X
```

```python
# Use the Function, inspect the output

S = my_function(Z,cellsize)
plt.imshow(S,cmap='bwr')
```

```python
# Save the output

with rasterio.open('output.tif', 'w', **profile) as dst:
    dst.write(S, 1)
```
