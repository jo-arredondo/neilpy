```python
import neilpy
import rasterio
import numpy as np
import scipy.ndimage as ndi
from skimage.util import apply_parallel
import matplotlib.pyplot as plt
import tifffile
```

# Load the data

```python
with rasterio.open('../sample_data/sample_dem.tif') as src:
        Z = src.read(1)
        Zt = src.transform
        cellsize = Zt[0]
```

# Slope

```python
S = neilpy.slope(Z,cellsize)
plt.imshow(S,'gray_r')
```

```python
Z,G = neilpy.rasterGi(S,footprint=7)
plt.imshow(G,'bwr',vmin=-3,vmax=3)
```

# Curvature

```python
C = ndi.filters.laplace(Z/cellsize)
plt.imshow(C,'bwr')
```

```python
Z,G = neilpy.rasterGi(C,footprint=7)
plt.imshow(G,'bwr',vmin=-3,vmax=3)
```

# Save the output

```python
plt.imsave('curvature_hotspots.png',G,cmap='bwr',vmin=-3,vmax=3)
tifffile.imsave('curvature_hotspots.tif',G)
neilpy.write_worldfile(Zt,'curvature_hotspots.tfw')
```
