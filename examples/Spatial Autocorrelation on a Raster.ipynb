```python
import neilpy
import rasterio
import numpy as np
import scipy.ndimage as ndi
from skimage.util import apply_parallel
import matplotlib.pyplot as plt
```

```python
with rasterio.open('../sample_data/sample_dem.tif') as src:
        Z = src.read(1)
        Zt = src.transform
        cellsize = Zt[0]
```

```python
S = neilpy.slope(Z,cellsize)
plt.imshow(S,'gray_r')
```

```python
Z,G = neilpy.rasterGi(S,footprint=7)
plt.imshow(G,'bwr',vmin=-3,vmax=3)
```
